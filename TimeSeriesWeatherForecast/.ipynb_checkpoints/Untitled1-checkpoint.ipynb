```python
def outlierdetect(columns):
    for i in columns:
        print("Column is : ",i)
        IQR = (np.percentile(df[i], 75)) - (np.percentile(df[i], 25))
        whisker_val = (np.percentile(df[i], 75)) + (1.5*(IQR))
        print(df.loc[df[i]>whisker_val].shape)
        print("================")
```

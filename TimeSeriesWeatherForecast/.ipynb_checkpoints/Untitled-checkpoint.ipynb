```python
import numpy as np
```

```python
IQR = (np.percentile(df['tempm'], 75)) - (np.percentile(df['tempm'], 25))
whisker_val = (np.percentile(df['tempm'], 75)) + (1.5*(IQR))
df.loc[df['tempm']>whisker_val].shape
```

```python
df['tempm'] = df['tempm'].apply(lambda x: np.nan if x > whisker_val else x)
```

```python
def rmsle(actual, preds):
    for i in range(0,len(preds)):
        if preds[i]<0:
            preds[i] = 0
        else:
            pass
    
    error = (sqrt(mean_squared_log_error(actual, preds)))*100
    return error
```

```python
from statsmodels.tsa.api import ExponentialSmoothing
#training the model
model = ExponentialSmoothing(np.asarray(train_data['tempm']), seasonal_periods=365, trend='add', seasonal='add')
model = model.fit(smoothing_level=0.2, smoothing_slope=0.001, smoothing_seasonal=0.2)
    
# predictions and evaluation
preds = model.forecast(len(valid_data)) 
score = rmsle(valid_data['tempm'], preds)

# results
print('RMSLE for Holt Winter is:', score)
```

```python
from itertools import product
from tqdm import tqdm_notebook

# setting initial values and some bounds for them
level = [0.1, 0.3, 0.5, 0.8]
smoothing_slope = [0.0001, 0.001, 0.05] 
smoothing_seasonal = [0.2, 0.4, 0.6]

# creating list with all the possible combinations of parameters
parameters = product(level, smoothing_slope, smoothing_seasonal)
parameters_list = list(parameters)
len(parameters_list)
```

```python
def grid_search(parameters_list):
    
    results = []
    best_error_ = float("inf")

    for param in tqdm_notebook(parameters_list):
        #training the model
        model = ExponentialSmoothing(np.asarray(train_data['tempm']), seasonal_periods=6, trend='add', seasonal='add')
        model = model.fit(smoothing_level=param[0], smoothing_slope=param[1], smoothing_seasonal=param[2])

        # predictions and evaluation
        preds = model.forecast(len(valid_data)) 
        score = rmsle(valid_data['tempm'], preds)
        
        # saving best model, rmse and parameters
        if score < best_error_:
            best_model = model
            best_error_ = score
            best_param = param
        results.append([param, score])
        
    result_table = pd.DataFrame(results)
    result_table.columns = ['parameters', 'RMSLE']
    
    
    # sorting in ascending order, the lower rmse is - the better
    result_table = result_table.sort_values(by='RMSLE', ascending=True).reset_index(drop=True)
    
    return result_table
```

```python
result_table = grid_search(parameters_list)
```

# Stationarity Test

```python
from statsmodels.tsa.stattools import adfuller
def adf_test(timeseries):
    
    #Perform Dickey-Fuller test:
    print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput=pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])

    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)
```

```python
adf_test(df['tempm'])
```

```python
def inverse_boxcox(y, lambda_):
    return np.exp(y) if lambda_ == 0 else np.exp(np.log(lambda_ * y + 1) / lambda_)
```

```python
from scipy import stats
train_data['tempm_log'], lambda_ar = stats.boxcox(train_data['tempm'])
lambda_ar
```

```python
train_data['tempm_log_diff']=train_data['tempm_log']-train_data['tempm_log'].shift(365)
```

```python
plt.figure(figsize=(12,8))

plt.plot(train_data.index, train_data['tempm_log_diff'])
plt.legend(loc='best')
plt.title("Stationary Series")
plt.show()
```

```python
train_data['tempm_log_diff_diff']=train_data['tempm_log_diff']-train_data['tempm_log_diff'].shift(1)
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plot_acf(train_data['tempm_log_diff_diff'].dropna(), lags=25)
plot_pacf(train_data['tempm_log_diff_diff'].dropna(), lags=25)
plt.show()
```

```python
from statsmodels.tsa.statespace import sarimax
model = sarimax.SARIMAX(train_data['tempm'], seasonal_order=(1,1,2,6), order=(1,1,2))
model = model.fit()
```

```python
end = len(train_data) + len(valid_data)
preds = inverse_boxcox(model.predict(start=0, end=end)[len(train_data)+1:], lambda_ar) 
preds = (preds.reset_index()) # storing validation predictions
```

```python
from sklearn.linear_model import LinearRegression

# separating features and target variable
x_train = train_data.drop(['tempm','Dates','time','heatindexm','conds','datetime_utc','datetimeutc_converted','wdire'], axis=1)
y_train = train_data['tempm']

x_valid = valid_data.drop(['tempm','Dates','time','heatindexm','conds','datetime_utc','datetimeutc_converted','wdire'], axis=1)
y_valid = valid_data['tempm']

#training the model
model = LinearRegression(normalize=True)
model.fit(x_train, y_train)
preds = model.predict(x_valid)

# results
score = rmsle(y_valid, preds)

print('RMSLE for Linear Regression is', score)
```

```python
feature_coeff = pd.DataFrame(zip(x_train.columns, model.coef_), columns=['Feature', 'coeff'])
feature_coeff
plt.bar(feature_coeff['Feature'], feature_coeff['coeff'])
```

```python
#regularization
```

```python
from sklearn.linear_model import Ridge
```

```python
for alpha in [0.01, 0.05, 0.1, 0.5, 1, 5]:
    print('----- ----- ----- ----- -----')
    print('At alpha value:', alpha)

    #training the model
    model = Ridge(alpha = alpha, normalize=True)
    model.fit(x_train, y_train)

    # predictions 
    preds = model.predict(x_valid)
    score = rmsle(y_valid, preds)

    # results
    print('Linear Regression error is', score)
```

```python
#training the model
model = Ridge(alpha = 0.5, normalize=True)
model.fit(x_train, y_train)

# predictions 
preds = model.predict(x_valid)
score = rmsle(y_valid, preds)

# results
print('Linear Regression error is', score)
```
